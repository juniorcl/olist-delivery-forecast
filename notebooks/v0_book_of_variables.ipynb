```python
def change_to_bold(text):

    return f'\033[1m{text}\033[0m'
```

```python
def get_data_frame_analysis(df):

    print(f"\n{change_to_bold('First 5 lines')}")
    display(df.head())

    print(f"\n{change_to_bold('Last 5 lines')}")
    display(df.tail())

    print(f"\n{change_to_bold('Types')}")
    display(df.dtypes)

    print(f"\n{change_to_bold('Shape')}")
    display(df.shape)

    print(f"\n{change_to_bold('Number of uniques')}")
    display(df.nunique())
    
    print(f"\n{change_to_bold('Percentage of NAs')}")
    display(df.isna().mean())

    print(f"\n{change_to_bold('Number of NAs')}")
    display(df.isna().mean())
```

```python
def get_distance_in_km(lat, lng, lat1, lng1):

    tuple_lat_lng = (lat, lng)
    tuple_lat1_lng1 = (lat1, lng1)
    
    try:
        dist = geodesic(tuple_lat_lng, tuple_lat1_lng1).km

    except:
        dist = np.nan

    return dist
```

```python
def get_categorical_mode(arr):
    
    """
    Returns the most frequent value in a numpy array.
    """
    
    arr = list(arr)
    
    return max(set(arr), key=arr.count)
```

```python
def turn_agg_into_dataframe(df: pd.DataFrame):

    columns = df.columns.values
    
    df.columns = ['_'.join(col).rstrip('_') for col in columns]
    df = df.reset_index()

    return df
```

## Distances of Customers and Sellers

```python
df_customer_geolocation = df_customers_order.loc[:, ['order_id', 'customer_id', 'customer_lat', 'customer_lng']]

df_customer_geolocation.head()
```

```python
df_order_seller_id = df_items.loc[:, ['order_id', 'seller_id']]
df_order_seller_id.head()
```

```python
aux = df_geolocation.drop_duplicates(
    subset='geolocation_zip_code_prefix').loc[:, [
    'geolocation_zip_code_prefix', 'geolocation_city',	'geolocation_state', 'geolocation_lat', 'geolocation_lng']]

df_sellers_geolocation = df_sellers.loc[:, ['seller_id', 'seller_zip_code_prefix']].merge(
    aux, 
    left_on='seller_zip_code_prefix', 
    right_on='geolocation_zip_code_prefix', 
    how='left').drop(columns=['geolocation_zip_code_prefix']).rename(columns={'geolocation_lat': 'seller_lat', 'geolocation_lng': 'seller_lng'})

df_sellers_geolocation.rename(columns={'geolocation_city': 'seller_geolocation_city', 'geolocation_state': 'seller_geolocation_state'}, inplace=True)

df_sellers_geolocation.head()
```

```python
df_distances = df_customer_geolocation.merge(df_order_seller_id, on='order_id', how='left')
df_distances = df_distances.merge(df_sellers_geolocation, on='seller_id', how='left')
df_distances = df_distances.loc[:, ['order_id', 'seller_geolocation_city',	'seller_geolocation_state', 'customer_lat', 'customer_lng', 'seller_lat', 'seller_lng']]
```

```python
df_distances.shape
```

```python
df_distances.head()
```

```python
df_distances['distance_customer_seller'] = df_distances.apply(
    lambda i: get_distance_in_km(i['customer_lat'], i['customer_lng'], i['seller_lat'], i['seller_lng']), axis=1)
```

```python
df_distances.isna().mean()
```

```python
df_distances.head()
```

```python
list_func_num_var = [
    'count', 
    'min', 
    'max', 
    'mean', 
    'median',
    [
        'q25', lambda i: np.quantile(i, 0.25)
    ],
    [
        'q75', lambda i: np.quantile(i, 0.75)
    ],
    [
        'range', lambda i: np.max(i) - np.min(i)
    ]
]
```

```python
list_func_cat_var = [
    'count', 
    'nunique',
    'size',
    [
        'mode', get_categorical_mode
    ]
]
```

```python
df_distances = df_distances.groupby('order_id').agg(
    {
        'seller_geolocation_city': list_func_cat_var,
        'seller_geolocation_state': list_func_cat_var,
        'distance_customer_seller': list_func_num_var        
    }
)

df_distances = turn_agg_into_dataframe(df_distances)
```

```python
df_distances.head()
```

```python
df_distances.shape
```

```python
df_distances.to_parquet('../data/interim/V0.0_dataframe_distances.parquet')
```

## Creating Variable from Date

```python
df.head()
```

```python
df['order_purchase_until_approved_in_seconds'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.total_seconds()
df['order_purchase_until_approved_in_minutes'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.total_seconds() / 60
df['order_purchase_until_approved_in_hours'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.total_seconds() / 3600
df['order_purchase_until_approved_in_days'] = (df['order_approved_at'] - df['order_purchase_timestamp']).dt.days
```

```python
df['order_purchase_timestamp_second'] = df['order_purchase_timestamp'].dt.second
df['order_purchase_timestamp_minute'] = df['order_purchase_timestamp'].dt.minute
df['order_purchase_timestamp_hour'] = df['order_purchase_timestamp'].dt.hour
df['order_purchase_timestamp_day'] = df['order_purchase_timestamp'].dt.day
df['order_purchase_timestamp_month'] = df['order_purchase_timestamp'].dt.month
```

```python
df['order_approved_at_second'] = df['order_approved_at'].dt.second
df['order_approved_at_minute'] = df['order_approved_at'].dt.minute
df['order_approved_at_hour'] = df['order_approved_at'].dt.hour
df['order_approved_at_day'] = df['order_approved_at'].dt.day
df['order_approved_at_month'] = df['order_approved_at'].dt.month
```

## Creating the Target

```python
df['delivered_in_days'] = (
    df['order_delivered_customer_date'] - df['order_approved_at']
).dt.days
```

```python
df['estimated_delivery_in_days'] = (
    df['order_estimated_delivery_date'] - df['order_approved_at']
).dt.days
```
